```python
# Import Modules
import os
import pandas as pd

#Find Current Path & List of Files/Folders in Current Path
current_path = os.getcwd()
current_directory = os.listdir()

#Loop Through List of Files/Folders in Current Path
for item in current_directory:
    
    #Set Condtion for Folder Containing School District Data
    if item.lower().find('school') >= 0 and item.lower().find('ipynb') == -1:

        #Loop Through List of Files/Folders in School District Data Folder
        for data in os.listdir(os.path.join(current_path, item)):
            
            #Set Condition for School Data CSV File
            if data.lower().find('school') >= 0 and data.lower().find('csv') >= 0:

                #Define Path Variable for School Data CSV File
                school_input = os.path.join(current_path, item, data)
                
            #Set Condition for Student Data CSV File
            elif data.lower().find('student') >= 0 and data.lower().find('csv') >= 0:

                #Define Path Variable for School Data CSV File
                student_input = os.path.join(current_path, item, data)

#Read School & Student CSV Files into Pandas Data Frames
school_data = pd.read_csv(school_input)
student_data = pd.read_csv(student_input)

#Merge School & Student Data Frames into Combined Data Frame
school_student_combined = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

#Display Merged Data Frame
school_student_combined
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate Total Number of Schools
total_schools = school_student_combined['school_name'].nunique()

#Caclculate Total Number of Students
total_students = school_student_combined['school_name'].count()

#Find Budgets for Each School & Calculate Total Budget
unique_budgets = school_student_combined['budget'].unique()

total_budget = unique_budgets.sum()

#Calculate Average Math & Reading Scores
math_avg = school_student_combined['math_score'].mean()
read_avg = school_student_combined['reading_score'].mean()

#Calculate Student Percentage Passing Math & Reading
math_pass = school_student_combined[school_student_combined['math_score'] >= 70]
read_pass = school_student_combined[school_student_combined['reading_score'] >= 70]

math_pct = ((math_pass['math_score'].count()) / total_students) * 100
read_pct = ((read_pass['reading_score'].count()) / total_students) * 100

#Calculate Overall Student Percentage Passing
overall_pass = (math_pct + read_pct) / 2

#Create Data Frame to Display District Summary
district_summary = pd.DataFrame({'Total Schools': [total_schools], 'Total Students': [format(total_students, ',d')],
                                 'Total Budget': ['${:,.2f}'.format(total_budget)],
                                 'Average Math Score': [round(math_avg, 3)],
                                 'Average Reading Score': [round(read_avg, 3)],
                                 '% Passing Math': [round(math_pct, 3)], '% Passing Reading': [round(read_pct, 3)],
                                 '% Overall Passing Rate': [round(overall_pass, 3)]})

#Display District Summary Data
district_summary
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
#Split Combined Data Frame into Groups by School Name
school_groups = school_student_combined.groupby('school_name')

#Create Data Frame to Display School Summary
school_summary = pd.DataFrame({'School Type': [], 'Total Students': [], 'Total Budget': [], 'Per Student Budget': [],
                               'Average Math Score': [], 'Average Reading Score': [], '% Passing Math': [],
                               '% Passing Reading': [], '% Overall Passing Rate': []})

#Initialize Variables for School Metric Data
school_name = []
school_type = []
total_students = []
total_budget = []
student_budget = []
math_avg = []
read_avg = []
math_pct = []
read_pct = []
overall_pass = []

#Loop Through Key-Value Pairs for School Names & Associated Data Groups
for group_name, group_data in school_groups:
    
    #Assign School Name to Variable
    school_name.append(group_name)
    
    #Find School Type
    school_class = group_data['type'].unique()
    
    school_type.append(school_class[-1])
    
    #Calculate Total Students
    total_students.append(group_data['school_name'].count())
    
    #Find Total School Budget
    budget_amount = group_data['budget'].unique()
    
    total_budget.append(budget_amount[-1])
    
    #Calculate Budget Per Student
    student_budget.append(total_budget[-1] / total_students[-1])
    
    #Calculate Average Math & Reading Scores
    math_avg.append(group_data['math_score'].mean())
    read_avg.append(group_data['reading_score'].mean())
    
    #Calculate Student Percentage Passing Math & Reading
    math_pass = group_data[group_data['math_score'] >= 70]
    read_pass = group_data[group_data['reading_score'] >= 70]

    math_pct.append(((math_pass['math_score'].count()) / total_students[-1]) * 100)
    read_pct.append(((read_pass['reading_score'].count()) / total_students[-1]) * 100)
    
    #Calculate Overall Student Percentage Passing
    overall_pass.append((math_pct[-1] + read_pct[-1]) / 2)
    
#Append Individual School Data to School Summary Data Frame
school_summary = pd.DataFrame({'School Type': school_type, 'Total Students': [format(item, ',d') for item in total_students],
                               'Total Budget': ['${:,.2f}'.format(item) for item in total_budget],
                               'Per Student Budget': ['${:,.2f}'.format(item) for item in student_budget],
                               'Average Math Score': [round(item, 3) for item in math_avg],
                               'Average Reading Score': [round(item, 3) for item in read_avg],
                               '% Passing Math': [round(item, 3) for item in math_pct],
                               '% Passing Reading': [round(item, 3) for item in read_pct],
                               '% Overall Passing Rate': [round(item, 3) for item in overall_pass]}, school_name)

school_summary
```

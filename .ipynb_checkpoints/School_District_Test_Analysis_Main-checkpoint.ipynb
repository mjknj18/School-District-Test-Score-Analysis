```python
# Import Modules
import os
import pandas as pd

#Find Current Path & List of Files/Folders in Current Path
current_path = os.getcwd()
current_directory = os.listdir()

#Loop Through List of Files/Folders in Current Path
for item in current_directory:
    
    #Set Condtion for Folder Containing School District Data
    if item.lower().find('school') >= 0 and item.lower().find('ipynb') == -1:

        #Loop Through List of Files/Folders in School District Data Folder
        for data in os.listdir(os.path.join(current_path, item)):
            
            #Set Condition for School Data CSV File
            if data.lower().find('school') >= 0 and data.lower().find('csv') >= 0:

                #Define Path Variable for School Data CSV File
                school_input = os.path.join(current_path, item, data)
                
            #Set Condition for Student Data CSV File
            elif data.lower().find('student') >= 0 and data.lower().find('csv') >= 0:

                #Define Path Variable for School Data CSV File
                student_input = os.path.join(current_path, item, data)

#Read School & Student CSV Files into Pandas Data Frames
school_data = pd.read_csv(school_input)
student_data = pd.read_csv(student_input)

#Merge School & Student Data Frames into Combined Data Frame
school_student_combined = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

#Display Merged Data Frame
school_student_combined
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate Total Number of Schools
total_schools = school_student_combined['school_name'].nunique()

#Caclculate Total Number of Students
total_students = school_student_combined['school_name'].count()

#Find Budgets for Each School & Calculate Total Budget
unique_budgets = school_student_combined['budget'].unique()

total_budget = unique_budgets.sum()

#Calculate Average Math & Reading Scores
math_avg = school_student_combined['math_score'].mean()
read_avg = school_student_combined['reading_score'].mean()

#Calculate Student Percentage Passing Math & Reading
math_pass = school_student_combined[school_student_combined['math_score'] >= 70]
read_pass = school_student_combined[school_student_combined['reading_score'] >= 70]

math_pct = ((math_pass['math_score'].count()) / total_students) * 100
read_pct = ((read_pass['reading_score'].count()) / total_students) * 100

#Calculate Overall Student Percentage Passing
overall_pass = (math_pct + read_pct) / 2

#Create Data Frame to Display District Summary
district_summary = pd.DataFrame({'Total Schools': [total_schools], 'Total Students': [format(total_students, ',d')],
                                 'Total Budget ($)': [round(total_budget, 2)],
                                 'Average Math Score': [round(math_avg, 3)],
                                 'Average Reading Score': [round(read_avg, 3)],
                                 '% Passing Math': [round(math_pct, 3)], '% Passing Reading': [round(read_pct, 3)],
                                 '% Overall Passing Rate': [round(overall_pass, 3)]})

#Display District Summary Data
district_summary
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
#Split Combined Data Frame into Groups by School Name
school_groups = school_student_combined.groupby('school_name')

#Create Data Frame to Display School Summary
school_summary = pd.DataFrame({'School Type': [], 'Total Students': [], 'Total Budget': [], 'Per Student Budget': [],
                               'Average Math Score': [], 'Average Reading Score': [], '% Passing Math': [],
                               '% Passing Reading': [], '% Overall Passing Rate': []})

#Initialize Variables for School Metric Data
school_name = []
school_type = []
total_students = []
total_budget = []
student_budget = []
math_avg = []
read_avg = []
math_pct = []
read_pct = []
overall_pass = []

#Loop Through Key-Value Pairs for School Names & Associated Data Groups
for group_name, group_data in school_groups:
    
    #Assign School Name to Variable
    school_name.append(group_name)
    
    #Find School Type
    school_class = group_data['type'].unique()
    
    school_type.append(school_class[-1])
    
    #Calculate Total Students
    total_students.append(group_data['school_name'].count())
    
    #Find Total School Budget
    budget_amount = group_data['budget'].unique()
    
    total_budget.append(budget_amount[-1])
    
    #Calculate Budget Per Student
    student_budget.append(total_budget[-1] / total_students[-1])
    
    #Calculate Average Math & Reading Scores
    math_avg.append(group_data['math_score'].mean())
    read_avg.append(group_data['reading_score'].mean())
    
    #Calculate Student Percentage Passing Math & Reading
    math_pass = group_data[group_data['math_score'] >= 70]
    read_pass = group_data[group_data['reading_score'] >= 70]

    math_pct.append(((math_pass['math_score'].count()) / total_students[-1]) * 100)
    read_pct.append(((read_pass['reading_score'].count()) / total_students[-1]) * 100)
    
    #Calculate Overall Student Percentage Passing
    overall_pass.append((math_pct[-1] + read_pct[-1]) / 2)
    
#Append Individual School Data to School Summary Data Frame
school_summary = pd.DataFrame({'School Type': school_type, 'Total Students': [format(item, ',d') for item in total_students],
                               'Total Budget ($)': [round(item, 2) for item in total_budget],
                               'Per Student Budget ($)': [round(item, 2) for item in student_budget],
                               'Average Math Score': [round(item, 3) for item in math_avg],
                               'Average Reading Score': [round(item, 3) for item in read_avg],
                               '% Passing Math': [round(item, 3) for item in math_pct],
                               '% Passing Reading': [round(item, 3) for item in read_pct],
                               '% Overall Passing Rate': [round(item, 3) for item in overall_pass]}, school_name)

#Sort School Summary Data Frames by Schools with Best Overall Passing Rate
best_schools = school_summary.sort_values('% Overall Passing Rate', ascending = False)

#Display Top 5 Schools with Best Overall Passing Rate
best_schools.head()
```

## Bottom Performing Schools (By Passing Rate)

* Sort and display the five worst-performing schools

```python
#Sort School Summary Data Frames by Schools with Worst Overall Passing Rate
worst_schools = school_summary.sort_values('% Overall Passing Rate', ascending = True)

#Display Top 5 Schools with Worst Overall Passing Rate
worst_schools.head()
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#Split Combined Data Frame into Groups by Grade Level
grade_groups = school_student_combined.groupby('grade')

#Create Data Frame to Display Average Math Score Summary
math_summary = pd.DataFrame({'9th Grade': [], '10th Grade': [], '11th Grade': [], '12th Grade': []})

#Initialize Variables for Math Score Metric Data
school_names = []
ninth_scores = []
tenth_scores = []
eleventh_scores = []
twelvth_scores = []

#Loop Through Key-Value Pairs for Grade Levels & Associated Data Groups
for grade_name, grade_data in grade_groups:
    
    #Split Grade Groups into Groups by School Name
    school_groups = grade_data.groupby('school_name')
    
    #Set Conditional Statements for Each Grade Level
    if grade_name == '9th':
        
        #Loop Through School Groups for 9th Grade Data
        for school_name, school_data in school_groups:
            
            #Create List of School Names
            school_names.append(school_name)
            
            #Extract & Average 9th Grade Math Scores & Add to 9th Grade Math Scores List
            ninth_scores.append(school_data['math_score'].mean())
    elif grade_name == '10th':
        
        #Loop Through School Groups for 10th Grade Data
        for school_name, school_data in school_groups:
            
            #Extract & Average 10th Grade Math Scores & Add to 10th Grade Math Scores List
            tenth_scores.append(school_data['math_score'].mean())
    elif grade_name == '11th':
        
        #Loop Through School Groups for 11th Grade Data
        for school_name, school_data in school_groups:
            
            #Extract & Average 11th Grade Math Scores & Add to 11th Grade Math Scores List
            eleventh_scores.append(school_data['math_score'].mean())
    elif grade_name == '12th':
        
        #Loop Through School Groups for 12th Grade Data
        for school_name, school_data in school_groups:
            
            #Extract & Average 12th Grade Math Scores & Add to 12th Grade Math Scores List
            twelvth_scores.append(school_data['math_score'].mean())
        
#Append Individual Grade Math Scores Data to Average Math Score Summary Data Frame
math_summary = pd.DataFrame({'9th Grade': [round(item, 3) for item in ninth_scores],
                             '10th Scores': [round(item, 3) for item in tenth_scores],
                             '11th Grade': [round(item, 3) for item in eleventh_scores],
                             '12th Grade': [round(item, 3) for item in twelvth_scores]}, school_names)

#Display Average Math Scores for Each School
math_summary
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#Set Bins for School Budget Per Student Ranges
spending_bins = [0, 585, 615, 645, 675]
spending_ranges = ['<$585', '$585-615', '$615-645', '$645-675']

#Cut School Summary Data Frame & Add Column for School Budget Per Student Ranges
school_summary['Per Student Spending Range ($)'] = pd.cut(school_summary['Per Student Budget ($)'], bins = spending_bins, labels = spending_ranges)

#Split School Summary Data Frame into Groups by School Budget Per Student Range
spending_groups = school_summary.groupby('Per Student Spending Range ($)')

#Create Data Frame to Display School Budget Per Student Range Summary
budget_summary = pd.DataFrame({'Average Math Score': [], 'Average Reading Score': [], '% Passing Math': [],
                             '% Passing Reading': [], '% Overall Passing Rate': []})

#Initialize Variables for School Budget Per Student Range Metric Data
math_avg = []
read_avg = []
math_pct = []
read_pct = []
overall_pass = []

#Loop Through Key-Value Pairs for School Budget Per Student Ranges & Associated Data Groups
for range_name, range_group in spending_groups:
    
    #Calculate Average Math Scores & Append to List
    math_avg.append(range_group['Average Math Score'].mean())
    
    #Calculate Average Reading Scores & Append to List
    read_avg.append(range_group['Average Reading Score'].mean())
    
    #Calculate Percent Passing Math & Append to List
    math_pct.append(range_group['% Passing Math'].mean())
    
    #Calculate Percent Passing Reading & Append to List
    read_pct.append(range_group['% Passing Reading'].mean())
    
    #Calculate Percent Overall Passing Rate & Append to List
    overall_pass.append(range_group['% Overall Passing Rate'].mean())
    
#Append Individual School Budget Per Student Average Range Data to School Budget Per Student Summary Data Frame
budget_summary = pd.DataFrame({'Average Math Score': [round(item, 3) for item in math_avg],
                               'Average Reading Score': [round(item, 3) for item in read_avg],
                               '% Passing Math': [round(item, 3) for item in math_pct],
                               '% Passing Reading': [round(item, 3) for item in read_pct],
                               '% Overall Passing Rate': [round(item, 3) for item in overall_pass]}, spending_ranges)

#Display Summary Data for School Per Student Budget Ranges
budget_summary
```
